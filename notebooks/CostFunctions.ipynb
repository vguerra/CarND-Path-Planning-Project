```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline 
```

```python
max_distance = 100
def closest_cost(x):
    return float(max_distance - min(x, max_distance))/max_distance

distances = range(0, 151)
costs = list(map(closest_cost, distances))

plt.xlabel('distance to closest car ahead')
plt.ylabel('cost')

plt.plot(np.array(distances), np.array(costs))
```

$$
CC(d) =
\begin{cases}
(100 - d)/100,  & \text{if $d$ is less or equal to 100} \\
0, & \text{otherwise}
\end{cases}
$$

```python
max_velocity=80
def fastest_cost(x):
    return float((max_velocity - min(x, max_velocity)))/max_velocity
    
velocities = range(0, 151)
costs = list(map(fastest_cost, velocities))

plt.xlabel('Velocity of car ahead')
plt.ylabel('cost')


plt.plot(np.array(velocities), np.array(costs))
```

$$
SC(v) =
\begin{cases}
(80 - d)/80,  & \text{if $d$ is less or equal to 80} \\
0, & \text{otherwise}
\end{cases}
$$

```python
car_radius=1.5
def colision_cost(x):
    colides = (x < 2 * car_radius)
    return 1.0 if colides else 0.0

distances = range(0, 20)
costs = list(map(colision_cost, distances))

plt.xlabel('distance to closest')
plt.ylabel('cost')


plt.plot(np.array(distances), np.array(costs))
```

$$
C(l) =
\begin{cases}
1,  & \text{if distance to closest car is less or equal to 2 * radius of car} \\
0, & \text{otherwise}
\end{cases}
$$

$$
Cost\_Lane(l) = W_{CL}*CL(l) + W_{CC}*CC(l) + W_{SC}*SC(l) + W_{C}*C(l)
$$

```python
def lane_change_cost(x):
    return pow(x, 2)
    
dist_to_target_lane = range(0, 3)
costs = list(map(lane_change_cost, dist_to_target_lane))

plt.xlabel('Count of lanes in change')
plt.ylabel('cost')

plt.xticks( np.arange(3))


plt.plot(np.array(dist_to_target_lane), np.array(costs))
```

$$
CL(l, l_{t}) = (l_{t} - l)^2
$$
