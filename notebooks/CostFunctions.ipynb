```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
def closest_cost(x):
    return float(100 - min(x, 100))/100

distances = range(0, 151)
costs = list(map(closest_cost, distances))
plt.plot(np.array(distances), np.array(costs))
```

```python
def fastest_cost(x):
    return float((100 - min(x, 100)))/100
    
velocities = range(0, 151)
costs = list(map(fastest_cost, velocities))
plt.plot(np.array(velocities), np.array(costs))
```
